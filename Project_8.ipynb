```python
df_trips = pd.read_csv('/datasets/project_sql_result_07.csv')
```

```python
df_trips.info()
df_trips.head()
```

```python
good = df_trips[df_trips['weather_conditions'] == 'Good']['duration_seconds']
```

```python
bad = df_trips[df_trips['weather_conditions'] == 'Bad']['duration_seconds']
```

<div class="alert alert-block alert-success"> <b>Comentario del revisor</b> <a class="tocSkip"></a><br>

Bien hecho!<b/>

La manera de separar la información de ambos días es correcta
</div>

```python
##H₀: La duración promedio de los viajes es la misma en sábados lluviosos y no lluviosos.

##H₁: La duración promedio de los viajes es diferente en sábados lluviosos.
```

```python
from scipy import stats
alpha = 0.05
```

```python
results = stats.ttest_ind(bad, good, equal_var=False)
results
```

```python
p_value = results.pvalue

if p_value < alpha:
    print('Rechazamos la hipótesis nula')
else:
    print('No rechazamos la hipótesis nula')

p_value
```

<div class="alert alert-block alert-success"> <b>Comentario del revisor</b> <a class="tocSkip"></a><br>

Bien hecho!<br/>

Es correcto rechazar la hipotesis nula, debido a que sí hay notables cambios entre un día de buenas condiciones y uno con malas condiciones
</div>

```python
##Ya que la hipotesis nula es rechazada, podemos decir que existe una evidencia estadistica sobre una mayor duracion de los viajes en dias de luvia
```

```python
mean_good = good.mean()
mean_bad = bad.mean()

print('Duración promedio con buen clima:', mean_good)
print('Duración promedio con mal clima:', mean_bad)
```

```python
##Como se puede ver aqui, la diferencia de tiempo es significativa, lo que para la empresa representa un incremento de ganancias
##Esto tambien nos podria indicar que las personas que de normal hubieran caminado, prefieren evitar la lluvia con ayuda de un taxi
##Con esta informacion, podemos recalcar la importancia de mantener la actividad de trabajo y reforzarla en dias de lluvia
```
